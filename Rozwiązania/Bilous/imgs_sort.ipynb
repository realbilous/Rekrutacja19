```python
import pandas as pd
import numpy as np
from PIL import Image, ImageStat
import shutil
import os
import math
```

```python
# Read csv, and Sort by Median
df = pd.read_csv('images.csv').sort_values("Median")
```

```python
for count, name in enumerate(list(df['Name'])):
    if not os.path.exists(f'agg-images/{math.floor(count / 4) +1}-images'):
        os.makedirs(f'agg-images/{math.floor(count / 4) +1}-images')
    shutil.copy2(f'images/{name}', f'agg-images/{math.floor(count / 4) +1}-images')
    
```
