```python
def find_coords(val):
    for k in range(grayscale_im.size[0]):   
       for j in range(grayscale_im.size[1]):  
          if np.array(pixel_arr[k,j]) == val:
                return [k, j];
                
```

```python
j = 0;
for f in os.listdir('.'):
    i = Image.open(f)
    fn, fext = os.path.splitext(f)
    #i.filename, i.size, " ".join(fn.split('-'))
    #pixel_arr = np.array(i)
    #sum1 = np.sum(pixel_arr, axis=0)
    #sum1 = np.sum(sum1, axis=0)
    #avg_color = tuple((sum1 / (pixel_arr.size / sum1.size)).astype(int))
    #print(avg_color)
    
    avg_color = tuple(np.array(ImageStat.Stat(i).mean).astype(int))
    print(avg_color)
    
    grayscale_im = i.convert('L')
    pixel_arr = np.array(grayscale_im)
    
    #median
    med = ImageStat.Stat(grayscale_im).median[0]
    print(med)
    
    min_max = grayscale_im.getextrema()
    print(min_max)
    
    coords = find_coords(min_max[1]);
    print(coords);
    j = j + 1;
    df.loc[f'{j}'] = [f"{i.filename}", f"{' '.join(fn.split('-')[:-1])}", f"{fn.split('-')[-1]}", f"{i.size[0]}", f"{i.size[1]}",
          f"{avg_color}", f"{med}", f"{coords[0]}", f"{coords[1]}"]
    

df.head()
```
